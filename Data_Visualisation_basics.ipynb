```python
semester = np.array([1,2,3,4,5])
marks = np.array([9.0, 8.5, 8.64, 8.64, 8.5])

x = np.array([1,2,3,4,5])
y = x ** 2 + x

plt.plot(x, y, label='Stock Price 1')
plt.plot(x, x ** 3 + 4, label='Stock Price 2')
plt.title('My first plot')
plt.legend()
plt.xlabel('Timestamp')
plt.ylabel('Stock Price')
plt.show()
```

```python
x = np.array([1,2,3,4,5])
y = x ** 2 + x

plt.scatter(x, y, label='Stock Price 1')
plt.scatter(x, x ** 3 + 4, label='Stock Price 2')
plt.title('My scatter plot')
plt.legend()
plt.xlabel('Timestamp')
plt.ylabel('Stock Price')
plt.show()
```
