# Feature Extraction algorithm (PCA)

## Principle Component Analysis

### Dimensionality Reduction
If no. of features is very large, our model will learn from the irrelevent/redundant data and perform bad. So, we have some techniques to reduce the number of features, namely -
#### 1) Feature Selection
#### 2) Feature Extraction

- PCA is one of the feature extraction algorithms.
- It is a dimensionality reduction technique.
- Reduce the features from 'n' dimensions to 'k' dimensions, where k < n
- eg: Convert 50D data to 5D/10D data
- PCA is the process of transferring a space having many dimensions into a space of fewer dimensions.

```python
import matplotlib.pyplot as plt
import cv2
img = cv2.imread('images/PCA1.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

### Application of PCA
- Data Compression
- Data Visualisation
- Speeds computation time

#### 1. Data Compression
- Reduce the number of dimensions in data from 50D(50 numbers to represent a vector/data point) -> 10D(10 numbers to represent a vector/data point)

##### Eg: We want to convert 2D data to 1D

```python
img = cv2.imread('images/PCA3.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

#### eg: 3D data -> 2D data

- We have to choose a plane
- And a plane has 2 features
- Project all points on the plane

```python
plt.figure(figsize=(10,4))
img = cv2.imread('images/PCA4.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

### IDEA

- IDea is that **most of the variance should be explained**
- Highest variance should be explained
- What is Variance? Variance is a spread
- Eg: We have only 2 features, x1 and x2, and x2 is always 5. It will not be relevent to consider x2. Just project all points on x1 axis and now this will be your new dataset
- Spread along y axis is very less  

```python
plt.figure(figsize=(10,4))
img = cv2.imread('images/PCA5.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

- Spread/Variance across y is 0
- Along x, spread is there
- Spread is actually your information
- **So, column which has more spread/variance is the more informative feature. That feature is explaining a lot of information. So, we cannot ignore that feature**

- When doing dimensionality reduction, **Most of the variance should be retained** or **data should be able to explain most of the variance**

## PCA OBJECTIVE

Steps in PCA-
- Data Preprocessing
- Covariance Matrix
- Find eigen vectors
- Project data onto new axis

### 1. Data Preprocessing
- Perform data standardisation (u = 0, std = 1) before applying PCA. 

- When converting to 1 dimension, we have to find a line
- When converting to 2 dimensions, we have to find a plane 

```python
plt.figure(figsize=(10,6))
img = cv2.imread('images/PCA7.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

- Project data points perpendicularly on the two red lines

```python
plt.figure(figsize=(10,6))
img = cv2.imread('images/PCA8.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

- In a), the spread is more and hence is explaining more data
- In b), Spread is very less
- The perpendicular distances are called **projection errors**
- We have to **minimise the sum of projection errors**

### OBJECTIVE IS - 'minimise the sum of projection errors' or 'maximize the spread'

```python
plt.figure(figsize=(10,6))
img = cv2.imread('images/PCA9.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

**Summary**
- To convert data from 2D to 1 dimension, find a vector(direction/axis) onto which to project the data so as to minimize the projection error
- To convert data from nD to k dimensions, find 'k' vectors(directions/axis) onto which to project the data so as to minimize the projection error

## PCA ALGORTIHM

Consider X_training_data  ->  (10000, 2)
- No. of training data points - 10000
- No. of features             - 2
- Convert from 2 dimension to 1 dimension

### STEP 1: Preprocessing (Standardization)
- u = 0, std = 0
- Can use the StandardScaler() while implementing
- First step was to find a direction (u) and then project data on this direction and these data points will be called z1
- x(1) - [x1, x2]
- z(1) - [z1]

- And this u (direction) must be such that the projection error is least (or the variance is high)

```python
plt.figure(figsize=(15,8))
img = cv2.imread('images/PCA10.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

- So we need to get this u vector
##### ALGORITHM
- Compute covariance matrix

### STEP 2: Compute Covariance matrix
- We know var(x) = (sum over m exmaples) ((x(i) - x.mean)^2) / N - 1
- And if we have 2 features, then covariance matrix is 2 x 2

```python
plt.figure(figsize=(15,8))
img = cv2.imread('images/PCA11.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

### STEP 3: Compute Eigen vectors of Covariance matrix
- |u| will be 1, as it will give the direction
- In the figure below, we have to find the direction vectors or the eigen vectors (u(1), u(2), u(3), ....)
- We use the function svd() - Singular Value Decomposition
- U, S, V = svd(covar)
- Here, U are the eigen vectors

##### Note: if we have X - (m, 5), then covariance matrix will be (5,5)

```python
plt.figure(figsize=(15,8))
img = cv2.imread('images/PCA12.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

- U.shape - (n, n)
- We take the first 'k' columns which are our eigen vectors

```python
plt.figure(figsize=(15,8))
img = cv2.imread('images/PCA13.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

- Ured = U[:, :k]
- Ured => (n x k)
- Now project the data points on the eigen vectors

### STEP 4: Projection
- Ured => (n x k)
- X    => (m x n)
- Final reduced matrix = np.dot( X , Ured ) => (m x k)

### Summary

```python
plt.figure(figsize=(15,8))
img = cv2.imread('images/PCA14.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()

plt.figure(figsize=(15,8))
img = cv2.imread('images/PCA15.JPG')
plt.axis('off')
plt.imshow(img)
plt.show()
```

```python
# Can also get explained_variance_ratio by
evr = pca1.explained_variance_ / np.sum(pca1.explained_variance_)
evr
```

```python
cum_explained_variance = np.cumsum(pca1.explained_variance_ratio_)
```

```python
cum_explained_variance
```

### The first number is telling that the first component is explaining 5.84% of the variance
### If we take 2 components, we are explaining - 10.127% variance only !!!

```python
plt.style.use('seaborn')
plt.plot(cum_explained_variance)
plt.xlabel('n_components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

### We can see that we can take only the first 200 components to explain 90% of the variance in data and

### we can take only the first 350 components to explain 95% of the variance in data
