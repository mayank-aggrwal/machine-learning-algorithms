### Line Plots

```python
x = np.arange(10)
y1 = x ** 2
y2 = 2 * x + 3

print(x, y1, y2)
```

```python
themes = plt.style.available
print(themes)
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(x, y1, label='Apple')
plt.plot(x, y2, label='Kiwi')
plt.title('Data Visualisation')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
plt.plot(x, y1, color='red', label='Apple', marker='o', linewidth=3)
plt.plot(x, y2, label='Kiwi', linestyle='dashed', marker='^')
plt.title('Data Visualisation')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

#### Plot indices on x-axis

```python
prices = np.array([1,2,3,4]) ** 3
plt.plot(prices)
```

### Scatter Plots

```python
x = np.arange(10)
y1 = x ** 2
y2 = 2 * x + 3

print(x, y1, y2)
```

```python
plt.style.use('seaborn')
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y1, color='red', label='Apple', marker='o', linewidth=0.4)
plt.scatter(x, y2, label='Kiwi', marker='^')
plt.title('Data Visualisation')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.show()
```

### Bar Graphs

```python
plt.bar?
```

```python
x = np.array([0,1,2])
plt.bar(x - 0.1, [20,10,14], label='Current Year', width=0.2, color='orange')
plt.bar(x + 0.1, [10,12,12], label='Next Year', width=0.2, tick_label=['Gold', 'Platinum', 'Silver'])
plt.title('Stock prices over the years')
plt.xlabel('Metal')
plt.ylabel('price')
plt.legend()
```

```python
plt.style.use('dark_background')
x = np.array([0,1,2])
plt.bar(x - 0.1, [20,10,14], label='Current Year', width=0.2, color='orange')
plt.bar(x + 0.1, [10,12,12], label='Next Year', color='green', width=0.2, tick_label=['Gold', 'Platinum', 'Silver'])
plt.title('Stock prices over the years')
plt.xlabel('Metal')
plt.ylabel('price')
plt.ylim(-5, 35)
plt.xlim(-1, 5)
plt.legend()
```

### Pie Charts

```python
plt.style.use('seaborn')
```

```python
weights = np.array([20,10,15,5])
subjects = ['Maths', 'Physics', 'Chemistry', 'English']
```

```python
plt.pie(weights, labels=subjects, explode=[0,0,0.1,0.3], autopct='%1.1f%%', shadow=True)
plt.show()
```

### Normal Distribution

```python
np.random.randn?
```

```python
Xsn = np.random.randn(100)
sigma = 5
u = 60

X = np.round(Xsn * sigma + u)
Xlist = list(np.round(Xsn * sigma + u))

X2 = np.round(Xsn * 7 + 70)

print(X)
```

### Histograms

```python
plt.style.use('seaborn')
plt.hist(Xlist, label='Physics')
plt.hist(X2, label='Maths', alpha=0.7)
plt.title('Marks vs Frequency histogram')
plt.xlabel('Marks')
plt.ylim(0, 35)
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
