```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/kick-2.jfif')
print(img.shape)
```

```python
plt.imshow(img)
plt.show()
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
# plt.axis('off')
plt.show()
```

```python
cropped_img = rgb_img[5:120, 55:170, :]
plt.imshow(cropped_img)
# plt.axis('off')
plt.show()
```

```python
resized_img = cv2.resize(cropped_img, (100, 100))
plt.imshow(resized_img)
plt.title('KICK - ' + str(resized_img.shape))
plt.show()
```

```python
resized_img = cv2.resize(cropped_img, (80, 80))
plt.imshow(resized_img)
plt.title('KICK - ' + str(resized_img.shape))
plt.show()
```

```python
resized_img = cv2.resize(cropped_img, (50, 50))
plt.imshow(resized_img)
plt.title('KICK - ' + str(resized_img.shape))
plt.show()
```
