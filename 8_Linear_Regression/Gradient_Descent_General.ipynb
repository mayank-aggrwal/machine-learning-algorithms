#### Gradient Descent

**Goal**
Given  a function F(X), we want to find the value of x that minimizes F(X)

```python
x = 0
lr = 0.1

# Take 50 steps in downhill direction
for i in range(50):
    grad = 2 * (x - 5)
    x = x - lr * grad
    print(x)

```

### Plot showing behaviour of Gradient Descent for different learning rates

```python
i = 1
plt.figure(figsize=(30, 10))

for j in range(0, 10, 1):
    
    X = np.arange(10)
    Y = (X - 5) ** 2
    
    x = 0
    lr = j / 10
    errors = []
    plt.subplot(2, 5, j + 1)
    plt.title('lr = ' + str(lr))
    plt.plot(X, Y)

    # Take 50 steps in downhill direction
    for i in range(50):
        grad = 2 * (x - 5)
        x = x - lr * grad
        y = (x - 5) ** 2
        errors.append(y)
        if i == 0:
            plt.scatter(x, y, color='Orange')
        else:
            plt.scatter(x, y, color='Red')
    
    
    i += 1
plt.show()
```
