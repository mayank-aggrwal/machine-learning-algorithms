## Data Visualisation of Movie dataset

Given a movie dataset, your task is to make a visualisation that -
- Plots the length of Movie Title names on X-axis
- Frequency counts on Y-axis

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('./Dataset/movie_metadata.csv')
```

```python
print(type(df))
print('Shape:', df.shape)
df.head(n=10)
```

```python
df.columns
```

```python
titles = df.get('movie_title')
print(titles)
print(type(titles))
```

```python
titles = list(titles)
print(titles)
```

```python
titles[0][:-1]
```

```python
for i in range(len(titles)):
    titles[i] = titles[i][:-1]
```

```python
print(titles)
```

```python
freq_map = {}
for title in titles:
    letters = len(title)
    if freq_map.get(letters) is None:
        freq_map[letters] = 1
    else:
        freq_map[letters] += 1
```

```python
print(freq_map)
```

```python
freq_map.keys()
```

```python
freq_map.values()
```

```python
X = freq_map.keys()
print(type(X))

X = list(X)
print(X)
```

```python
Y = freq_map.values()
print(type(Y))

Y = list(Y)
print(Y)
```

```python
plt.style.use('seaborn')
plt.scatter(X, Y, color='red')
plt.title('Movie title length vs frequency')
plt.xlabel('Movie title length')
plt.ylabel('Frequency')
plt.show()
```
